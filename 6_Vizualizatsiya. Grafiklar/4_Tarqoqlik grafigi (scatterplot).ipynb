##Scatter plot - Tarqoqlik diagrammasi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df=df[df.year>2000]
plt.figure(figsize=(16,10))
sns.scatterplot(data=df, x='mileage',y='price', hue='transmission',s=100)
#sns.regplot(data=df, x='mileage',y='price')
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'],y=df['price'])
plt.show()
```

```python
sns.lmplot(data=df, x='price',y='mileage',row='transmission',row='model')
plt.show()
```
