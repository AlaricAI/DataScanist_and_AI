```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010]
df=df[df.transmission!='Other']
df=df[df.fuelType!="Other"]
df.head()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
fig1=plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(x,sin,
         linewidth=1.5,
         linestyle='-.',
         color='y',
         alpha=0.7,
         marker='h')
plt.title('sin(x) grafigi')
plt.grid(color='tomato',alpha=0.5, linestyle='--')
plt.subplot(2,1,2)
plt.plot(x,cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')
plt.title('sin(x) va cos(x) funksiya grafigi')
plt.grid(color='tomato',alpha=0.5, linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,10))
fig.suptitle("Grafik 2")
sns.histplot(ax=ax[0,0], data=df,x='price')
sns.histplot(ax=ax[0,1], data=df,x='mileage')
sns.scatterplot(ax=ax[1,0], data=df, x=df['year'],y=df['price'],hue='fuelType', s=50)
sns.scatterplot(ax=ax[1,1], data=df, x=df['year'],y=df['mileage'],hue='fuelType', s=50)
ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price va Year distribution")
ax[1,1].set_title("Mileage va Year distribution")
plt.show()
```
