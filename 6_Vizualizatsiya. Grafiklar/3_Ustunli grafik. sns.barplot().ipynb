#Ustunli deagrammalar

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns= ['maydoni','aholisi']
df
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['maydoni'])
plt.title("O'zbekiston Respublikasi tarkibidagi hududlar aholi soni")
plt.xlabel("Hudud")
plt.ylabel("Maydoni kv.km")
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
df['month']=[i.month for i in df.index]
df.head()
```

```python
df['year']=[i.year for i in df.index]
df.head()
```

```python
df2=df.groupby(['month','year']).max()
df2.head()
```

```python
df2.unstack(level=0)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(level=0),cmap="Blues", linewidths=1, square=True)
xtic=['Jan','Feb','Mar','Apr','May','Jun',
      'Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(np.arange(12) + .5,labels=xtic)
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel("Minutes")
plt.show()
```

```python
melted_df = pd.melt(df, id_vars=['Month'], value_vars=['AA', 'MQ'], var_name='Airline', value_name='Delay')

# Yonma-yon solishtirish uchun ustunli diagramma
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Delay', hue='Airline', data=melted_df)
plt.title("Aviakompaniyalar bo'yicha o'rtacha kechikish vaqti")
plt.ylabel("O'rtacha kechikish vaqti (Minutes)")
plt.show()
```

```python
sns.catplot(data=melted_df, kind="bar", x="Month", y="Delay", hue="Airline")
plt.show()
```

```python
data = pd.melt(df, id_vars=['Month'], value_vars=['AA', 'AS', 'DL', 'WN', 'MQ'])
data.head()
```

```python
sns.catplot(data=data, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=df.set_index('Month'), cmap="Accent", annot=True)
plt.xlabel("Airline")
plt.show()
```
