## Ierarxik (hierarchical) algoritm

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%204/data/cars_clus.csv"
df = pd.read_csv(url)
df.head()
```

```python
df[[ 'sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']] = df[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']].apply(pd.to_numeric, errors='coerce')
df = df.dropna()
df = df.reset_index(drop=True)
df.head(5)
```

```python
featureset = df[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

```python
from sklearn.preprocessing import MinMaxScaler
x = featureset.values # numpy array
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx[0:5]
```

```python
import scipy
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
leng = feature_mtx.shape[0]
D = np.zeros([leng,leng])
for i in range(leng):
    for j in range(leng):
        D[i,j] = scipy.spatial.distance.euclidean(feature_mtx[i], feature_mtx[j])
D
```

```python
Z = hierarchy.linkage(D)
```

```python
fig = plt.figure(figsize=(18,50))
def llf(id):
    return '[%s %s %s]' % (df['manufact'][id], df['model'][id], int(float(df['type'][id])) )

dendro = hierarchy.dendrogram(Z, leaf_label_func=llf, leaf_rotation=0, leaf_font_size=12, orientation='right')
```

```python
from sklearn import manifold, datasets
from sklearn.cluster import AgglomerativeClustering
```

```python
agglom = AgglomerativeClustering(n_clusters = 6, linkage = 'complete')
agglom.fit(D)
agglom.labels_
```

```python
df['cluster'] = agglom.labels_
agg_cars = df.groupby(['cluster','type'])[['horsepow','engine_s','mpg','price']].mean()
agg_cars
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Ma'lumot yaratamiz: 2 ta oy shaklidagi cluster
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# DBSCAN modelini yaratamiz: eps=0.2, min_samples=5
model = DBSCAN(eps=0.2, min_samples=5)
labels = model.fit_predict(X)

# Natijani chizamiz
plt.figure(figsize=(8, 5))
plt.title("DBSCAN natijasi (eps=0.2, min_samples=5)")
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=60, edgecolors='k')
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors

# 5-ta yaqin qoshni nuqta uchun masofa
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

# Masofalarni osish tartibida chizamiz
sorted_distances = np.sort(distances[:, 4])  # 4-chi ustun - 5-chi yaqin masofa
plt.figure(figsize=(8,5))
plt.plot(sorted_distances)
plt.title("k-distance graph (5-chi yaqin qoshni)")
plt.xlabel("Nuqtalar")
plt.ylabel("Masofa")
plt.grid(True)
plt.show()
```

```python
# Har bir klasterda nechta nuqta bor?
from collections import Counter

cluster_counts = Counter(labels)
print("Klasterlar va ularning o'lchami:")
for label, count in sorted(cluster_counts.items()):
    print(f"Klaster {label}: {count} ta nuqta")
```

```python
np.random.seed(42)
latitudes = np.random.uniform(40.5, 41.0, size=100)
longitudes = np.random.uniform(69.2, 69.3, size=100)

gps_data = np.column_stack((latitudes, longitudes))

dbscan_gps = DBSCAN(eps=0.02, min_samples=5)
labels_gps = dbscan_gps.fit_predict(gps_data)

plt.figure(figsize=(8, 5))
plt.title("DBSCAN bilan GPS koordinatalarini klasterlash")
plt.scatter(gps_data[:, 0], gps_data[:, 1], c=labels_gps, cmap='rainbow', s=60, edgecolors='k')
plt.xlabel("Kenglik (Latitude)")
plt.ylabel("Uzunlik (Longitude)")
plt.grid(True)
plt.show()
```

```python
import hdbscan

hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=5)
labels_hdbscan = hdbscan_model.fit_predict(gps_data)

plt.figure(figsize=(8, 5))
plt.title("HDBSCAN bilan GPS koordinatalarini klasterlash")
plt.scatter(gps_data[:, 0], gps_data[:, 1], c=labels_hdbscan, cmap='rainbow', s=60, edgecolors='k')
plt.xlabel("Kenglik (Latitude)")
plt.ylabel("Uzunlik (Longitude)")
plt.grid(True)
plt.show()
```
