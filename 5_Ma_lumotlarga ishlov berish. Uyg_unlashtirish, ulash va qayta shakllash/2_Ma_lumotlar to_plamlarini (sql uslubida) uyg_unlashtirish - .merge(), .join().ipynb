```python
A=pd.DataFrame({'Transport':['poezd','avtomobil','kema'],
                'Tezligi':['70','180','45']})
A
```

```python
B=pd.DataFrame({'Transport':['poezd','avtomobil','tramvay'],
                'Narxi':['98000','10000','65000']})
B
```

```python
merged = pd.merge(A, B,how='inner')
merged
```

```python
merged = pd.merge(A,B, how='outer')
merged
```

```python
merger = pd.merge(A, B, how='left')#left bu chapga birlashish
merger
```

```python
merger = pd.merge(A, B, how='right')#right bu o'ng birlashish
merger
```
